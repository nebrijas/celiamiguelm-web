```python
import pandas as pd
```

```python
url = 'https://api.covid19api.com/countries'
```

Se hace una petición a la API y el resultado se guarda en una variable.

```python
df = pd.read_json(url)
```

```python
df
```

Para seleccionar tan solo la fila requerida, se hace un filtrado por país, en este caso España.

```python
df[df['Country'] == 'Spain']
```

## Datos en tiempo real del Covid19 en España

Se utiliza el mismo método para obtener los datos de la API. Con la librería pandas, se llama a la funcion read_json(), que devuelve los datos de la consulta.

```python
url_rt_es = 'https://api.covid19api.com/country/spain/status/confirmed/live'
df_rt_es = pd.read_json(url_rt_es)
df_rt_es
```

Método para obtener las 5 primeras filas.

```python
df_rt_es.head()
```

Método para obtener las 5 últimas filas.

```python
df_rt_es.tail()
```

Primero se definen la "x" y la "y" para crear el gráfico. Después se muestra con la función plot().

```python
casos_es = df_rt_es.set_index('Date')['Cases']
casos_es.plot(title="Covid19 en España")
```

## Datos en tiempo real del Covid19 en Panamá

```python
df[df['Country'] == 'Panama']
```

```python
url_rt_pa = 'https://api.covid19api.com/country/panama/status/confirmed/live'
df_rt_pa = pd.read_json(url_rt_pa)
df_rt_pa
```

```python
casos_pa = df_rt_pa.set_index('Date')['Cases']
casos_pa.plot(title="Covid19 en Panamá")
```

## Diferencias de casos Covid19 entre España y Panamá

El primer paso es unir los casos de ambos países en una misma tabla. Para eso se utiliza la funcion concat().

```python
pa_vs_es= pd.concat([casos_es, casos_pa], axis=1)
pa_vs_es
```

Se define el nombre de la columna con el país correspondiente.

```python
pa_vs_es.columns = ('España', 'Panamá')
pa_vs_es
```

```python
pa_vs_es.plot(title= "Covid19 España vs Panama")
```
